```python
import kagglehub
import zipfile
import os
import pandas as pd

# 1. Download dataset (this will be a zip file)
zip_path = kagglehub.dataset_download("arnabchaki/popular-video-games-1980-2023")
print("Downloaded zip path:", zip_path)

# 2. Extract the zip
extract_folder = "video_games_dataset"
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_folder)

# 3. Find the CSV file
csv_files = [f for f in os.listdir(extract_folder) if f.endswith('.csv')]
print("CSV files found:", csv_files)

# 4. Load CSV
df = pd.read_csv(os.path.join(extract_folder, csv_files[0]))
print("Dataset shape:", df.shape)
```
